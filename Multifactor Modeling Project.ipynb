```python
import statsmodels.api as sm
import numpy as np
import pandas as pd

# Create a list of the style portfolio returns
returns = ['S/L', 'S/M', 'S/H', 'B/L', 'B/M', 'B/H']

# Create a list of the risk factors
risk_factors = ['RMRF', 'ΔTerm Spread', 'ΔYield Spread', '%Δ in Oil Price']

# Create an empty list to store the regression results
results = []

# Loop through the style portfolio returns
for r in returns:
    # Define the independent and dependent variables
    X = df[risk_factors]
    y = df[r]
    
    # Fit the linear regression model
    model = sm.OLS(y, X).fit()
    
    # Store the regression results
    results.append(model)

# Extract the coefficients from the results
coefficients = [result.params.values for result in results]

# Create the coefficient matrix
coefficient_matrix = np.array(coefficients).reshape(6, 4)

# Create a DataFrame from the coefficient matrix
coefficient_df = pd.DataFrame(coefficient_matrix, index=returns, columns=risk_factors)

# Print the coefficient DataFrame
print(coefficient_df)
```

```python
# Import the numpy library
import numpy as np

# Multiply the covariance matrix with the coefficient dataframe
new_matrix = np.dot(coefficient_df,covariance_matrix)

# Print the new matrix
print(new_matrix)

```

```python

# Multiply the new matrix with the transpose of the coefficient dataframe
Lo_FV_Lo_Matrix = np.dot(new_matrix, coefficient_df.T)

# Print the new matrix
print(Lo_FV_Lo_Matrix)
```

```python
# Add the new matrix 2 with the residual dataframe
return_covariance_matrix = Lo_FV_Lo_Matrix + residual_df

# Print the new matrix
print(return_covariance_matrix)

```

```python
# Calculate the mean of the T-Bill returns
t_bill_mean = df['T-Bill'].mean()

# Create a list to store the Rp values
Rp = []

# Loop through the style portfolio returns
for r in returns:
    # Calculate the mean of the style portfolio returns
    portfolio_mean = df[r].mean()
    
    # Calculate Rp by subtracting the T-Bill mean from the portfolio mean
    rp = portfolio_mean - t_bill_mean
    
    # Append the Rp value to the list
    Rp.append(rp)

# Print the Rp values
print(Rp)

```

```python
import cvxpy as cp

# Set the number of style portfolios
n = len(returns)

# Create variables for the portfolio weights
w = cp.Variable(n)

# Set the objective function to maximize the portfolio return
objective = cp.Maximize(w.T * Rp)

# Set the constraints
constraints = [cp.sum(w) == 1, w >= 0]

# Solve the optimization problem
prob = cp.Problem(objective, constraints)
result = prob.solve()

# Print the optimal portfolio weights
print(f'Optimal portfolio weights: {w.value}')
```
